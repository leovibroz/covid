### Arrumando nome dos paises

```python
import math
```

```python
new_names = {
        'United States of America':'US',
        'Russian Federation':'Russia',
        'United Kingdom':'UK'
        
}
```

```python
pop_3['country_region'] = pop_3.apply(lambda x: new_names[x['country_region']] if  x['country_region'] in new_names  else x['country_region'],axis = 1)
```

## Cruzando casos com populacao

```python
cas_pop_0 = pd.merge(cas_1,pop_3,how='left',on=['country_region'])
```

```python
cas_pop_0[cas_pop_0['country_code'].isnull()]['country_region'].unique()
```

```python
cas_pop_0['current_date'] = pd.to_datetime(cas_pop_0['current_date'])
```

# Analises

## Taxa de mortalidade

```python
cas_pop_0['tx_mort'] = cas_pop_0['deaths']/(cas_pop_0['total_pop']/1000000)
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
plot_test = cas_pop_0[cas_pop_0['country_region'].isin(['Brazil','US','UK','France','Italy'])]
fig = px.line(plot_test, x="current_date", y="tx_mort", color='country_region')
fig.show()
```

## Data que chegou a 1 morte por milhao

```python
prim_1_morte= cas_pop_0[cas_pop_0['tx_mort'] >= 1].groupby(['country_region']).agg({'current_date':'min'})
prim_1_morte.columns = ['prim_1_morte']
```

```python
cas_pop_1 = pd.merge(cas_pop_0,prim_1_morte['prim_1_morte']
                     ,how='left',left_on=['country_region'],left_index = False,right_index = True)
```

```python
cas_pop_1['prim_1_morte'] = pd.to_datetime(cas_pop_1['prim_1_morte'])
```

```python
cas_pop_1['dias_desde_1_morte'] = cas_pop_1.apply(lambda row: (row['current_date'] - row['prim_1_morte']).days,axis = 1)
```

```python
cas_pop_1[cas_pop_1['country_region'] == 'Brazil']
```

```python
plot_test = cas_pop_1[cas_pop_1['country_region'].isin(['Brazil','US','UK','France','Italy'])]
fig = px.line(plot_test, x="dias_desde_1_morte", y="tx_mort", color='country_region')
fig.show()
```
