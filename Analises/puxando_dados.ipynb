```python
df_pop = pd.read_csv('Analises/dados_origem/populacao.csv',encoding='ISO-8859-1',delimiter = ';')
```

```python
df_pop.columns
```

```python
df_pop.columns = ['index', 'variant', 'country_region', 'Notes',
       'country_code', 'type', 'parent_code', 'dt_ref',
       '0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39',
       '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75-79',
       '80-84', '85-89', '90-94', '95-99', '100+']
```

```python
df_pop['type'].unique()
```

```python
df_pop_0 = df_pop[df_pop['type'] == 'Country/Area']
```

```python
int_col = ['0-4', '5-9', '10-14',
       '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54',
       '55-59', '60-64', '65-69', '70-74', '75-79', '80-84', '85-89', '90-94',
       '95-99', '100+']
```

```python
def convert_int(x):
    if type(x) == str:
        return  int(x.strip().replace(' ',''))*1000
    else:
        return x
```

```python
for col in int_col:
    df_pop_0[col] = df_pop_0[col].apply(lambda x: convert_int(x))
    df_pop_0[col] = pd.to_numeric(df_pop_0[col])
```

```python
df_pop_0.to_csv('Analises/pop_country.csv')
```
